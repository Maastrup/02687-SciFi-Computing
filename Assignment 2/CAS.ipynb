```python
def lmm_coeff(i,j):
    if i == 0:
        if j < 3:
            return 1
        else: 
            return 0
    elif j < 3:
        return 1/ factorial(i) * j ** i
    else:
        return -1/factorial(i - 1) * (j-3) ** (i-1)

A = lambda n,m: Matrix(n,m, lmm_coeff)
A(5,6)
```

```python
a = symbols('a')
alpha_beta = Matrix([a, -(1 + a), 1, -(1 + a) / 2, (3-a) / 2, 0])
alpha_beta
```
